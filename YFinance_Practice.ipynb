```python
msft = yf.Ticker("MSFT")
```

```python
hist = msft.history(period="max")
```

```python
#Gives high and lows for each day
hist
```

```python
# Dividends or stock splits
msft.actions

# msft.splits
# msft.dividends
```

```python
# Gives distribution of stock holders
msft.major_holders
```

```python
# Gives distribution of the companies that own microsoft stock
msft.institutional_holders
```

```python
#Sustainability Information
msft.sustainability
```

```python
# The major firms' opinions on the status of the microsoft stock price
msft.recommendations
```

```python
# The company's earning report
msft.calendar
```

More Advanced Financial Data Gathering

```python
# Major Indicies
major_indices = pd.read_html("https://finance.yahoo.com/markets/world-indices")[0]
```

```python
major_indices.head()
```

```python
ticker_list = major_indices['Symbol'].str.replace("^","").str.lower().to_list()
```

```python
df = yf.download(ticker_list, period="1d", start = "2020-01-13", end="2025-01-06")
```
